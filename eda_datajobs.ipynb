## NaN values and duplicates

```python
df_ds.isna().sum()
```

```python
df_ds.duplicated().sum()
```

```python
df_ds.drop_duplicates(inplace=True)
```

```python
df_ds.sort_values(by=['work_year','job_title'], inplace=True)
```

```python
df_ds.reset_index(inplace=True,drop=True)
```

```python
df_ds.columns
```

```python
df_ds = df_ds[['work_year','job_title','employment_type','experience_level','salary_in_usd','remote_ratio','employee_residence','company_location','company_size']]
```

## Fixing values

Employment Type:
* PT Part-time 
* FT Full-time 
* CT Contract 
* FL Freelance


```python
df_ds.employment_type.value_counts()
```

```python
employment_type = {
    'PT': 'Part-time',
    'FT': 'Full-time',
    'CT': 'Contract',
    'FL': 'Freelance'
}
```

```python
# Replacing employment_type categories
for find, repl in employment_type.items():
    df_ds.employment_type = df_ds.employment_type.str.replace(find, repl)
```

```python
df_ds.employment_type.value_counts()
```

Experience Level:
* EN Entry-level / Junior 
* MI Mid-level / Intermediate 
* SE Senior-level / Expert 
* EX Executive-level / Director

```python
df_ds.experience_level.value_counts()
```

```python
experience_level = {
    'EN': 'Junior',
    'MI': 'Semi-Senior',
    'SE': 'Senior',
    'EX': 'Executive'
}
```

```python
# Replacing experience_level categories
for find, repl in experience_level.items():
    df_ds.experience_level = df_ds.experience_level.str.replace(find, repl)
```

Company Size:
* S: Small (0-50 employees)
* M: Medium (50-250 employees)
* L: Large (more than 250 employees)

```python
df_ds.company_size.value_counts()
```

```python
company_size = {
    'S': 'Small',
    'M': 'Medium',
    'L': 'Large',
}
```

```python
# Replacing company_size categories
for find, repl in company_size.items():
    df_ds.company_size = df_ds.company_size.str.replace(find, repl)
```
